# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import os

# Files to load
city_data_to_load = os.path.join("Resources", "city_data.csv")
ride_data_to_load = os.path.join("Resources", "ride_data_3.csv")
#print(ride_data_to_load)
```

```python
# Read the city data file and store it in a pandas DataFrame.
city_data_df = pd.read_csv(city_data_to_load)
city_data_df.head(10)
```

```python
# Read the ride data file and store it in a pandas DataFrame.
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head(10)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
# Create the Urban city DataFrame.
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
urban_cities_df.head()
```

```python
# Create the Suburban cities DataFrames.
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
suburban_cities_df
```

```python
# Create the Rural city DataFrames.
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
rural_cities_df
```

```python
# 1. Get the total rides for each city type
total_rides_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_by_type.head
```

```python
# 2. Get the total drivers for each city type
total_drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_by_type.head()
```

```python
#  3. Get the total amount of fares for each city type
total_fare_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fare_by_type.head()
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride = total_fare_by_type /total_rides_by_type
average_fare_per_ride.head()
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = total_fare_by_type / total_drivers_by_type
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_ride_summary_df = pd.DataFrame({
    "Total Rides": total_rides_by_type,
    "Total Drivers": total_drivers_by_type,
    "Total Fares": total_fare_by_type,
    "Average Fare per Ride": average_fare_per_ride,
    "Average Fare per Driver": average_fare_per_driver})
pyber_ride_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_ride_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_ride_summary_df["Total Fares"] = pyber_ride_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_ride_summary_df["Average Fare per Ride"] = pyber_ride_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_ride_summary_df["Average Fare per Driver"] = pyber_ride_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
pyber_ride_summary_df["Total Rides"] = pyber_ride_summary_df["Total Rides"].map("{:,}".format)
pyber_ride_summary_df["Total Drivers"] = pyber_ride_summary_df["Total Drivers"].map("{:,}".format)
pyber_ride_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_of_fares = pyber_data_df.groupby(["type", "date"]).sum()["fare"]
sum_of_fares.head(10)
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
sum_of_fares = sum_of_fares.reset_index()
sum_of_fares.head(10)
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
sum_of_fares=sum_of_fares.pivot(index="date", columns="type", values="fare") 
sum_of_fares.head(10)
```

```python
sum_of_fares.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
sum_fares_Jan_through_Apr_df = sum_of_fares.loc['2019-01-01':'2019-04-29'] 
sum_fares_Jan_through_Apr_df

```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
sum_fares_Jan_through_Apr_df.index = pd.to_datetime(sum_fares_Jan_through_Apr_df.index)
sum_fares_Jan_through_Apr_df
#sum_fares_Jan_through_Apr_df = sum_fares_Jan_through_Apr_df.reset_index()
#sum_fares_Jan_through_Apr_df.head()
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
sum_fares_Jan_through_Apr_df.info()

sum_fares_Jan_through_Apr_df.index = pd.to_datetime(sum_fares_Jan_through_Apr_df.index)
sum_fares_Jan_through_Apr_df.tail

sum_fares_Jan_through_Apr_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
sum_fares_Jan_through_Apr_weekly_df = sum_fares_Jan_through_Apr_df.resample("W").sum()
print(sum_fares_Jan_through_Apr_weekly_df.shape)
sum_fares_Jan_through_Apr_weekly_df.head(40)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# Create the plot with ax.plt()
ax = sum_fares_Jan_through_Apr_weekly_df.plot(figsize = (20,6))

ax.set_title("Total Fare by City Type")
ax.set_xlabel("Month")
ax.set_ylabel("Fare")
ax.legend(loc="upper right")
plt.savefig("analysis/Challenge_fare_summary.png")
plt.show()
# Set the x-axis to a list of strings for each month.


#plt.plot(x_axis, y_axis)
```
